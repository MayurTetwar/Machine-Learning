# MICE using sklearn

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import warnings
warnings.filterwarnings("ignore")
```

```python
df = np.round(pd.read_csv(file_path)[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
df = df.iloc[:,0:-1]
temp = df.head().copy()
temp.head()
```

```python
df['R&D Spend'][0] = np.nan
df['Administration'][2] = np.nan
df['Marketing Spend'][4] = np.nan
```

```python
df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Using Random Forest inside MICE (this mixture of random forest with MICE called as MissForest)
# This is powerful for non-linear relationships!
mice_forest = IterativeImputer(
    estimator=RandomForestRegressor(n_jobs=-1,n_estimators=10), # The "brain" of the imputer
    max_iter=10,
    random_state=0
)
data_imputed_forest = mice_forest.fit_transform(df)
```

```python
df_final = pd.DataFrame(data_imputed_forest, columns=list(df.columns))
print("Data after MICE Imputation:")
temp-df_final.head().round(2)
```
