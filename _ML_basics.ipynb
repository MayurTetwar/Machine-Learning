# Machine Learning Fundamentals for Data Analysts
---

## 1. Supervised Learning
*Learning with a teacher (Labeled Data)*

### **A. Classification (Predicting Categories)**
* **Goal:** Assign data to a discrete label.
* **Examples:** * Email: Spam or Not Spam.
  * Customer: Will churn or stay.
  * Image: Dog or Cat.

### **B. Regression (Predicting Numbers)**
* **Goal:** Predict a continuous numerical value.
* **Examples:** * Predicting house prices.
  * Forecasting monthly sales revenue.
  * Estimating the temperature.

## 2. Unsupervised Learning
*Learning without a teacher (Unlabeled Data)*

* **Clustering:** Grouping similar data points together.
    * *Example:* Segmenting customers by purchasing behavior.
* **Dimensionality Reduction:** Reducing the number of features while keeping essential info.
    * *Example:* Compressing 100 columns of data into 3 main "components" (PCA).
* **Anomaly Detection:** Finding the "odd one out."
    * *Example:* Detecting fraudulent bank transactions.
* **Association Rule Learning:** Finding relationships between variables.
    * *Example:* Market Basket Analysis (People who buy bread also buy butter).

## 3. Semi-Supervised Learning
*The Best of Both Worlds*
* Uses a **small amount of labeled data** and a **large amount of unlabeled data**.
* **Use Case:** When labeling data is too expensive (e.g., a doctor manually labeling 10,000 X-rays).

## 4. Reinforcement Learning (RL)
*Learning by Experience (Reward & Penalty)*
* An **Agent** takes actions in an **Environment** to maximize a **Reward**.
* **Use Case:** Self-driving cars, Game AI (Chess/AlphaGo), Robotics.
